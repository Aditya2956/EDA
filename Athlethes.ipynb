```python
#Flitering down to all USA races in 2020 of length 50km/ 50mi
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2['Athlete Age'] = 2020-df2['Athlete year of birth']
```

```python
df2.head(10)
```

```python
df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df2.isna().sum()
```

```python
df2[df2['Athlete Age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
df2.dtypes
```

```python
df2 = df2.rename( columns  = {
        'Year of event' : 'year',
        'Event dates' : 'day',
        'Event name' : 'event_name',
        'Event distance/length' : 'distance',
        'Event number of finishers' : 'race_finishers',      
        'Athlete performance': 'performance',
        'Athlete gender' :  'gender',
        'Athlete average speed':    'speed',    
        'Athlete ID' :   'athelete_id',
        'Athlete Age':  'age' 
})
```
